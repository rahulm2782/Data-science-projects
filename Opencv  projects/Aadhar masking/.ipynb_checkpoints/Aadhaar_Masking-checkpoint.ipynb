# <font color='cyan'>AADHAR MASKING USING OCR

### <font color='cyan'>Importing necessary libraries

```python
import cv2
import re
import easyocr
from pytesseract import Output
import matplotlib.pyplot as plt
reader = easyocr.Reader(['en'])
```

```python
# Reading the image file from the location
img=cv2.imread("aadhar-card-900-16509875443x2.jpg") 

gray_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY) #image to grayscale
```

```python
#visualizing the aadhar card
image = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(20,15))
plt.imshow(image)
plt.axis('off')
plt.show()
```

```python
# reading the grayscaled using easy ocr
data1 = reader.readtext(gray_img)
```

<b><font color='cyan'>we know that aadhar has a length of 14 units when spaces are also considered.

```python
cord = [] #to store the cordinates of the aadhar numbers recored

for i in data1:
    if i[-1]>.50 and len(i[-2])==14: #if confidence > 50% and length of text == 14
        cord.append(i[0])
print(cord)
```

```python
len(cord)
```

<b><font color='cyan'> if len of cord greater than 2 that is we detected multiple aadhar no.s, we have 
    to mask them both

```python
xmin,ymin = [min(i) for i in zip(*cord[0])]   #origin point of the bounding box
x1min,y1min = [min(i) for i in zip(*cord[1])] #final width and length 
```

```python
xmax,ymax = [max(i) for i in zip(*cord[0])]
x1max,y1max = [max(i) for i in zip(*cord[1])]
```

### <b><font color='cyan'> Drawing Bounding boxes using cv2

<b><font color='cyan'> <b> to mask the aadhar we have to block the visibility of the first 8 digits

<b><font color='cyan'> we have total 3 sets of 4 digit number we create a new variable x to seperate the last 4 digits such that we can apply mask to the first 8 digits 

```python
x = int((xmax-xmin)/3) 
```

```python
x
```

<b><font color='cyan'> Similarly creating x1 for second bounding box

```python
x1 = int((x1max-x1min)/3)
```

```python
x1
```

<b><font color='cyan'>  Drawing a opaque box around the boundary, thickness=-1

```python
cv2.rectangle(img,(xmin,ymin),((xmax-x),ymax),(0,0,0),-1)
cv2.rectangle(img,(x1min,y1min),((x1max-x1),y1max),(0,0,0),-1)
plt.figure(figsize=(20,15))
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.axis('off')
```
